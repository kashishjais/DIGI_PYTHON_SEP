```python
from PIL import Image
```

```python
we=Image.open('grls.jpg')
doraemon=Image.open('img.jpg')
we
```

```python
w,h=we.size
```

```python
we.resize((500, 500))
```

```python
resized_we= we.resize((w//4, h//4))
resized_we
```

```python
Image.blend(resized_we,doraemon.resize((w//4, h//4)), 0.4)
```

```python
cake= Image.open('dora.png')
Image.blend(cake.resize((800, 400)).convert('RGB'),doraemon.resize((800, 400)), 0.7)
```
